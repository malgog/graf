```python
import random
#zad1
def random_graph(n, p):
    graph = {i: [] for i in range(n)}
    for i in range(n):
        for j in range(i + 1, n):
            if random.random() < p:
                graph[i].append(j)
                graph[j].append(i)
    return graph
print(random_graph(6,0.5))
```

```python
#zad2
def graph_to_matrix(graph):
    vertices = list(graph.keys())
    index = {v: i for i, v in enumerate(vertices)}
    n = len(vertices)

    matrix = [[0] * n for _ in range(n)]

    for v in vertices:
        for u in graph[v]:
            i = index[v]
            j = index[u]
            matrix[i][j] = 1

    return vertices, matrix
```

```python
#zad3
def matrix_to_graph(vertices, matrix):
    n = len(vertices)
    graph = {v: [] for v in vertices}

    for i in range(n):
        for j in range(n):
            if matrix[i][j] == 1:
                graph[vertices[i]].append(vertices[j])

    return graph
```

```python
#zad4
def cycle(n):
    graph = {i: [] for i in range(n)}

    for i in range(n):
        graph[i].append((i + 1) % n)
        graph[i].append((i - 1) % n)

    return graph

print(cycle(6))
```
